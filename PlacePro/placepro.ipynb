```python
object_columns = [
    'Knows ML', 'Knows DSA', 'Knows Python', 'Knows JavaScript', 
    'Knows HTML', 'Knows CSS', 'Knows Cricket', 'Knows Dance', 
    'Participated in College Fest', 'Was in Coding Club'
]

# Map 'Yes' to 1 and 'No' to 0 for these columns
for col in object_columns:
    data[col] = data[col].map({'Yes': 1, 'No': 0})
```

```python
data.head()
```

```python
data.head()
```

```python
#now working on null values
data.isnull().sum()
```

```python
# all the columns like Knows ML,Knows Python etc have null values and each of them has numeric as datatype so replacing each column's null value by it's mean
data["Knows ML"].fillna(int(data["Knows ML"].mean()), inplace=True)
data["Knows Python"].fillna(int(data["Knows Python"].mean()), inplace=True)
data["Knows JavaScript"].fillna(int(data["Knows JavaScript"].mean()), inplace=True)
data["Knows HTML"].fillna(int(data["Knows HTML"].mean()), inplace=True)
data["Knows CSS"].fillna(int(data["Knows CSS"].mean()), inplace=True)
```

```python
data.isnull().sum()
```

```python
#visualising the dataset
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig=plt.figure(figsize=(29,29))
correlation_matrix=data.corr(numeric_only=True)
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.show()
```

```python
#as Knows Dance,Knows Cricket,Participated in College Fest, and Interview Room Temperature are correlated very less in our dataset so removing them from our dataset
data = data.drop(columns=["Knows Dance", "Knows Cricket", "Interview Room Temperature"])
```

```python
#now visualizing my dataset through plotting some plots
package_counts = data["Placement Package"].value_counts()
plt.figure(figsize=(10, 6))  # Set figure size
plt.bar(package_counts.index, package_counts.values, color='skyblue')
plt.xlabel("Placement Package", fontsize=12)
plt.ylabel("Number of Students", fontsize=12)
plt.title("Number of Students Achieving Different Placement Packages", fontsize=14)
```

```python
branch_counts = data["Branch of Engineering"].value_counts()
plt.figure(figsize=(10, 5))  # Set figure size
plt.bar(branch_counts.index, branch_counts.values, color='skyblue')
plt.xlabel("Branch of Engineering", fontsize=12)
plt.ylabel("No of students belonging to a branch", fontsize=12)
```

```python
#Branch of Engineering also has a object as datatype so for converting it into numerical value we will use:
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
data['Name of Student']=label_encoder.fit_transform(data['Name of Student'])
data['Branch of Engineering'] = label_encoder.fit_transform(data['Branch of Engineering'])
```

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split,RandomizedSearchCV
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=1/3)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train, y_train)
```

```python
regressor.coef_
```

```python
print(regressor.intercept_)
```

```python
print(regressor.score(x,y)*100)
```
